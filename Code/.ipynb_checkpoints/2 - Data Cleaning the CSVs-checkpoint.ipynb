#### After downloading a bunch of articles from my webscraper. This notebook will detail my data cleaning process to create one nice CSV at the end.

#### Right now, I have mutliple CSVs and I need to get rid of Null values, duplicates, and any unnessary information that I do not need. The essentials only.

```python
# Pandas will be the main python tool I'll need
# Since I will be handling the cleaning with DataFrames
import pandas as pd

# The glob module finds all the pathnames matching a specified pattern
import glob
```

```python
# will create a simple function to prepare my csv files to LDA Topic Modeling later on. 


def clean_df_for_lda(DataFrame):
    '''
    Function for cleaning the CSVs that I created for my specific webscraper
    '''
    
    df_to_clean = pd.DataFrame()
    df_to_clean = DataFrame
    
    
    # Drop Authors and Published Data column
    # I had these columsn in my webscraper but I determined I didn't really need it 
    # during my explatory analysis 
    df_to_clean = df_to_clean.drop(['Unnamed: 0','Authors', 'published_date','Summary','Keywords'], axis = 1)
    
    #Drop any any null rows 
    df_to_clean = df_to_clean.dropna(axis = 0)
    
    #Drop any duplicated rows, since my web scraper was a little bit buggy
    df_to_clean = df_to_clean.drop_duplicates(subset='Text', keep = 'first')
    
    return df_to_clean
    
```

```python
path = 'data/unclean_data/webscraped/*.csv'

total_len_df = []

list_dfs = [] 

# For loop to read all the csv file and put into a DataFrame 
for csv_file_name in glob.glob(path):
    print(csv_file_name)
    df = pd.read_csv(csv_file_name )
    list_dfs.append(df)
    
    print(len(df))
    total_len_df.append(len(df))
    
```

```python
# To check the loop worked correctly with the same number of rows
sum(total_len_df)
```

```python
final_df = pd.concat(list_dfs, ignore_index= 'True')
print(final_df.shape)
```

```python
final_df.head()
```

```python
# Send the big DataFrame for cleaning
final_df_clean = clean_df_for_lda(final_df)
#reset the index
final_df_clean = final_df_clean.reset_index(drop = True)
```

```python
final_df_clean.to_csv('data/clean_scraped_data.csv')
```
