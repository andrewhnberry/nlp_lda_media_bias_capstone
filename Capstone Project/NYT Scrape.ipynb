```python
nyt_df = pd.DataFrame()

for i in np.arange(1,6):
    
    url = f'https://api.nytimes.com/svc/archive/v1/2019/{i}.json?api-key=nYc8XwaFDU8Sn3eOJMDdDFnxqhY305uL'
    
    r = requests.get(url)
    json_data = r.json()
    num_docs = pyjq.all('.response .docs | length',json_data)[0]
    jq_query = f'.response .docs [] | {{Title: .headline .main, Author: .byline .original, Summary: .snippet,  published_date: .pub_date, Source: .source, url: .web_url}}'
    output = pyjq.all(jq_query, json_data)
    
    t_df = pd.DataFrame(output)
    nyt_df = nyt_df.append(t_df, ignore_index = True)
    
```
